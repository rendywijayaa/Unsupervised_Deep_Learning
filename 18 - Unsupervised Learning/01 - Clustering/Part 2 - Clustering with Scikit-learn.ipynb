```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering, DBSCAN

from luwiji.cluster import demo
```

# Sample Data

```python
X1, y1 = demo.blob_data()
X2, y2 = demo.moon_data()
X3, y3 = demo.circle_data()
```

# Visualize

```python
X = X1
plt.scatter(X[:,0], X[:,1], s=10)
plt.axis("equal");
```

# K-means clustering

```python
X = X1
kmeans = KMeans(n_clusters=4)
member = kmeans.fit_predict(X)
center = kmeans.cluster_centers_

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c=member, s=10)
plt.scatter(center[:,0], center[:,1], c="w", edgecolors="k", s=50, linewidths=2)
plt.axis("equal");
```

# K-means weakness
* kesulitan untuk menentukan K
    - Solusi: elbow analysis, silhouette score, and others
* Hanya untuk data yang bisa dirata-ratakan
    - Solusi: K-medoid, K-prototype
* Solusi tidak selalu optimal
* Hanya untuk data yang linearly-separable

## Contoh: Moon Dataset

```python
X = X2
kmeans = KMeans(n_clusters=2)
member = kmeans.fit_predict(X)
center = kmeans.cluster_centers_

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c=member, s=10)
plt.scatter(center[:,0], center[:,1], c="w", edgecolors="k", s=50, linewidths=2)
plt.axis("equal");
```

## Contoh: Donat Dataset

```python
X = X3
kmeans = KMeans(n_clusters=2)
member = kmeans.fit_predict(X)
center = kmeans.cluster_centers_

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c=member, s=10)
plt.scatter(center[:,0], center[:,1], c="w", edgecolors="k", s=50, linewidths=2)
plt.axis("equal");
```


# Menentukan K (n_cluster) tanpa label -> Inertia / Elbow Analysis

```python
X = X1
score=[]
for k in range(2,25):
    kmeans = KMeans(n_clusters=k)
    member = kmeans.fit_predict(X)
    score.append(kmeans.inertia_)

plt.figure(figsize=(10,5))
plt.plot(range(2,25), score, "r-")
plt.ylabel("Inertia")
plt.xlabel("n_clusters");

# lihat elbow nya. ada di angka 4. berarti clusternya ada 4. 
```

# Menentukan K (n_cluster) dan ada label -> Silhouette score

```python
from sklearn.metrics import silhouette_score
```

```python
X = X1
score=[]
for k in range(2,25):
    kmeans = KMeans(n_clusters=k)
    member = kmeans.fit_predict(X)
    
    siluet = silhouette_score(X, member, metric='euclidean')
    score.append(siluet)

plt.figure(figsize=(10,5))
plt.plot(range(2,25), score, "r-")
plt.ylabel("Silhouette")
plt.xlabel("n_clusters");

# cari titik yang paling tinggi (peak nya), arahnya ke 4 cluster
```
